```python
atlantic_hurr['Category'] = atlantic_hurr['Category'].str.strip().str.upper()
pacific_hurr['Category'] = pacific_hurr['Category'].str.strip().str.upper()
```

```python
def Category_Deliniation(row):
    print(f"Processing: {row['Category']}, Wind Speed: {row['Highest Wind Speed']}")  # Debugging line
    if row['Category'] == "HU":
        if 74 <= row['Highest Wind Speed'] < 95:
            print("Assigning: Category 1")
            return 'Category 1'
        elif 95 < row['Highest Wind Speed'] < 110:
            print("Assigning: Category 2")
            return 'Category 2'
        elif 110 <= row['Highest Wind Speed'] < 130:
            print("Assigning: Category 3")
            return 'Category 3'
        elif 130 <= row['Highest Wind Speed'] < 156:
            print("Assigning: Category 4")
            return 'Category 4'
        elif row['Highest Wind Speed'] >= 156:
            print("Assigning: Category 5")
            return 'Category 5'
    return row['Category']  # Return original for non-HU
```

```python
atlantic_hurr['Category'] = atlantic_hurr.apply(Category_Deliniation, axis=1)
```

```python
atlantic_hurr.head(10)
```

```python
pacific_hurr['Category'] = pacific_hurr.apply(Category_Deliniation, axis=1)
```

```python
pacific_hurr.head(10)
```

```python
Hurricane_Data = pd.concat([pacific_hurr, atlantic_hurr])
```

```python
print(Hurricane_Data[Hurricane_Data['Year']==1924])
```

```python
Hurricane_Data.shape
```

```python
# Group by 'ID' and aggregate the data
final_hurr_data = Hurricane_Data.groupby('ID', as_index=False).agg({
    'Highest Wind Speed': 'max',   # Average the 'highest_windspeed'
    'Year': 'first',               # Take the first value of 'year' for each ID
    'Category': 'first'            # Take the first value of 'category' for each ID
})

# Rename the 'highest_windspeed' column
final_hurr_data.rename(columns={'Highest Wind Speed': 'average_windspeed'}, inplace=True)
final_hurr_data['average_windspeed'] = final_hurr_data['average_windspeed'].round()
```
